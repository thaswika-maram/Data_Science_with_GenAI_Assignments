# Python basics questions

1. What is python and why is it popular?

2. What is an interpreter in python?
- When you run a Python script, the Python interpretor translates your human-readable code into machine code on the fly, one line at a time, and executes it immediately.

3. What are pre-defined keywords in python?

7. What is the difference between "==" and "is" operator in python?
- In Python, == and is are not the same — they serve different purposes:
- == (Equality Operator):
     - Checks if two variables have the same value.
     - Compares contents of objects.
- is (Identity Operator):
     - Checks if two variables point to the same object in memory.
     - Compares object identity (memory address).

9. What is typecasting in python?
- Implicit Type Casting (Automatic)
     - Python automatically converts one data type to another without user involvement.
     - Happens when mixing different numeric types (e.g., int with float).
- Explicit Type Casting (Manual)
     - You manually convert data types using Python's built-in functions.

10. What is the difference between implicit and explicit type casting ?
-  Implicit Type Casting
    - Implicit type casting occurs automatically when Python converts one data type to another during expression evaluation.
    -  This usually happens when performing operations involving mixed types, like integers and floats. Python promotes the data type to prevent data loss.
     - For example, when you add an integer and a float, Python automatically converts the integer to a float before performing the operation. This process is safe and handled internally by the interpreter, so the programmer doesn't need to do anything.
- Explicit Type Casting
     - Explicit type casting is when the programmer manually converts one data type to another using built-in functions like int(), float(), str(), etc. This is useful when you're dealing with user input (which comes as a string) or when you need to convert types deliberately for a specific purpose.
     - Because you're forcing a conversion, you must ensure it's valid—otherwise, Python will raise an error (e.g., trying to convert "abc" to an integer will cause a ValueError).


11. What is the purpose of conditional statements in Python ?
- Conditional statements in Python are used to make decisions in a program based on certain conditions. They allow the program to execute different blocks of code depending on whether a condition is True or False.
- Control Program Flow:
    - They let you control the order in which code runs, making your program dynamic and responsive to different inputs or situations.
- Decision Making:
     - They help the program decide what to do next based on values, inputs, or computations.
- Add Logic to Programs:
    - You can implement logical operations like comparisons, choices, or checks (e.g., login validation, menu selection, data filtering, etc.).



12. How does the elif statement work ?
- The elif statement in Python is short for "else if" and is used to check multiple conditions in sequence. It comes after an if statement and allows you to check for additional conditions if the initial if condition is false.
    - The if statement checks the first condition.
    - If the if condition is True, it executes the block of code associated with it.
    - If the if condition is False, the program moves to the elif block (if any), checking the next condition.
    - If none of the if or elif conditions are True, the program will execute the else block (if provided).


13. What is the difference between for and while loops ?
- The for and while loops in Python are both used for iteration, but they differ in how they handle the repetition and the conditions under which they run. 
- For loop:
    - A for loop is generally used when you know how many times you want to literate or when you're working with a collection (like a list, tuple, or string.)
    - Iterable can be range, list, tuple, string, etc.
    - The loop iterates over each element in the iterable one by one.
- While loop:
    - A while loop is ussed when you want to repeat a block of code as long as a certain condition is True. The loop continues until the condition becomes False.
    - The loop runs as long as the condition is True.
    - If the condition is False initially, the code inside the loop won't run at all.

14. Describe a scenario where a while loop is more suitable than a for loop ?
- A while loop is more suitable than a for loop when you don’t know in advance how many times the loop should run, and the loop depends on a condition that might change in an unpredictable way during execution.
- example scenario:
     - Suppose you're writing a program that asks the user to enter a password, and you want to keep prompting them until they enter the correct one. You don't know how many attempts it will take, so a while loop is ideal.
     - Another scenario, In many games, the main game loop continues to run until the player decides to quit, which makes a while loop the perfect choice.

```python
# 8.Write a program to demonstrate the use of logical operators.

age = int(input("Enter your age: "))
has_id = input("Do you have a valid ID? (yes/no): ".lower())
# Logical AND: must be 18+ AND have ID to enter
if age >= 18 and has_id == "yes":
    print("Access granted.")
else:
    print("Access denied.")
# Logical OR: either age is under 12 OR has no ID = minor
if age < 18 or has_id == "no":
    print("You are considered a minor or undocumented.")
# Logical NOT: reversing a condition
if not has_id == "yes":
    print("You must carry a valid ID next time.")
```

```python
# 9.Write a Python program to convert user input from string to integer, float, and boolean types

user_input = input("Enter a value: ")

# Convert to integer
try:
    int_value = int(user_input)
    print(f"Integer conversion: {int_value}")
except ValueError:
    print("Cannot convert to integer.")

# Convert to float
try:
    float_value = float(user_input)
    print(f"Float conversion: {float_value}")
except ValueError:
    print("Cannot convert to float.")

# Convert to boolean
# In Python, any non-empty string is True; empty string is False
bool_value = bool(user_input)
print(f"Boolean conversion: {bool_value}")
```

```python
# 10.Write code to demonstrate type casting with list elements

string_list = ["10", "20", "30", "40"]
# Convert each string element to integer using list comprehension
int_list = [int(x) for x in string_list]
# Convert each integer to float
float_list = [float(x) for x in int_list]
# Convert each number back to string
back_to_string_list = [str(x) for x in float_list]
# Convert to boolean (non-zero numbers become True)
bool_list = [bool(x) for x in int_list]
# Print results
print("Original string list:", string_list)
print("After converting to integers:", int_list)
print("After converting to floats:", float_list)
print("After converting back to strings:", back_to_string_list)
print("After converting to booleans:", bool_list)
```

```python
# 14.Write a program to reverse a string using a while loop

text = input("Enter a string: ")
#initialize variables
reversed_text = ""
index = len(text) - 1
# Loop backwards through the string
while index >= 0:
    reversed_text += text[index]
    index -= 1
# Print the reversed string
print("Reversed string:", reversed_text)
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num = int(input("Enter a number to calculate its factorial: "))
# Initialize variables
factorial = 1
i = 1
# Calculate factorial using a while loop
while i <= num:
    factorial *= i
    i += 1
# Print the result
print(f"The factorial of {num} is: {factorial}")
```
